# pandas series


```python
import numpy as np
S = pd.Series([4.5,8,9,7,9,10])
print(S)
print(type(S))
```

```python
S.name = "this is a number"
print(S)
print(type(S))
```

```python
S.values
```

```python
print(S[0])
print(S[2])
```

```python
S.index
```

```python
S.index = {
    'num1',
    'num2',
    'num3',
    'num4',
    'num5',
    'num6'
}
print(S)
```

```python
S = pd.Series({
    'num1' : 100,
    'num2' : 101,
    'num3' : 102,
    'num4' : 103,
    'num4' : 104
}, name = 'this is a number')
print(S)
print("-------")
print(S['num1'])
```

```python
print(S.iloc[1])
print("-------")
print(S.iloc[-1])
```

```python
print(S[{'num1', 'num2'}])
print("-------")
print(S.iloc[[0,1]])
print("-------")
print(S['num1' : 'num3'])
```

```python
print(S * 100)
print("-------")
print(S > 102)
print("-------")
print(S[S > 101])
```

```python
print(S.mean())
print("-------")
print(S[S > S.mean()])
```

```python
print(S.std())
print(S[S > S.std()])
print("-------")
```

```python
print(S[(S > S.mean() - S.std() / 2) | (S > S.mean() + S.std() / 2)])
# | ----> or
```

```python
np.log(S)
```

```python
print(S[(S > 100) & (S < 102)])
#& --> and
```

```python
print(S[(S > 100) | (S < 102)])
# | ----> or
```

```python
S['num1'] = 500
print(S)
print("-------")
S[S > 101] = 600
print(S)
```
